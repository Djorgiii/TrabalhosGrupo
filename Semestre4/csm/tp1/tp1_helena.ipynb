### Exercicio 8

```python
# a imagem é de n*n, logo o centro da imagem é o pixel (n/2, n/2)
def imagem_setores(N, angulo_sector):
    
    img = np.zeros((N, N))

    cx = N // 2
    cy = N // 2

    for y in range(N):
        for x in range(N):

            dx = x - cx
            dy = y - cy

            ang = np.degrees(np.arctan2(dy, dx))

            if ang < 0:
                ang += 360

            sector = int(ang // angulo_sector)

            if sector % 2 == 0:
                img[y, x] = 0   # preto
            else:
                img[y, x] = 255 # branco
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()

imagem_setores(800, 1)
```
